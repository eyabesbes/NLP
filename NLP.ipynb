```python
import pandas as pd
col_list = ['pattern','question','answer']
df_train= pd.read_csv("/content/sample_data/NLP.txt",encoding='utf-8',sep=',',usecols=col_list)
df_train

```

```python
df_train['pattern'] = df_train['pattern'].astype(str)
df_train['question'] = df_train['question'].astype(str)
df_train['answer'] = df_train['answer'].astype(str)
```

Suppressionn des données manquantes:

```python
df_train.dropna(inplace=True)
```

verfication de la syntaxe:

```python
from textblob import TextBlob

def correct_sentence(sentence):
    return str(TextBlob(sentence).correct())

# Apply the function to the 'sentence' column of df_train
df_train['pattern'] = df_train['pattern'].apply(correct_sentence)
df_train.pattern

```

Suppression de la ponctuation:

```python
import string

# Remove punctuation from the 'pattern' column
def remove_punctuations(text):
    # Define the set of Arabic punctuation marks to remove
    arabic_punctuations = "،؛؟«»،؛؟٬٫٬٫"
    english_punctuations =string.punctuation
    # Remove the punctuation marks from the text
    return text.translate(str.maketrans('', '', arabic_punctuations + english_punctuations))

#df_train['pattern'] = df_train['pattern'].apply(lambda x: x.translate(str.maketrans("","", string.punctuation)))
df_train['removed_punctuation_answer'] = df_train['answer'].apply(remove_punctuations)
df_train['removed_punctuation_question'] = df_train['question'].apply(remove_punctuations)
#df_train['question'] = df_train['question'].apply(lambda x: x.translate(str.maketrans("","", string.punctuation)))

df_train
```

Suppression des stopwords:

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def remove_stopwords(text):
    stopwords_list = stopwords.words('arabic')
    tokens = text.split()
    filtered_tokens = [word for word in tokens if word.lower() not in stopwords_list]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text
df_train['filtered_question'] = df_train['removed_punctuation_question'].apply(remove_stopwords)
df_train['filtered_answer'] = df_train['removed_punctuation_answer'].apply(remove_stopwords)
```

```python
#df_train['question'] = df_train['question'].apply(correct_sentence)
#df_train.question
```

```python
from nltk.stem.isri import ISRIStemmer

st = ISRIStemmer() # Stemmer function
def stemm(sentence):
  l1=[]
  L=sentence.split()
  for i in L:
    s=st.stem(i)
    l1.append(s)
  return l1

df_train['question_stemmer']=[stemm(x) for x in df_train['filtered_question']]
df_train['answer_stemmer']=[stemm(x) for x in df_train['filtered_answer']]
df_train
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD
import random
import tensorflow as tf
```

```python
from tensorflow.keras.optimizers import SGD

# Create model - 3 layers. First layer 128 neurons, second layer 64 neurons, and 3rd output layer contains the number of neurons
# equal to the number of intents to predict output intent with softmax
model = Sequential()
model.add(Dense(128, input_shape=(len(df_train['vectorized_question'][0]),), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(df_train['vectorized_answer'][0]), activation='softmax'))

# Compile model. Stochastic Gradient Descent with Nesterov accelerated gradient gives good results for this model
sgd = SGD(learning_rate=0.05, nesterov=True)  # Use learning_rate instead of lr
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

# Fitting and saving the model
hist = model.fit(X.toarray(), y.toarray(), epochs=200, batch_size=50, verbose=1)
model.save('chatbot_model.h5', hist)
print("Model created")
```

```python
#Import transformers
import transformers


#Load arabert preprocessor
from arabert.preprocess import ArabertPreprocessor
arabert_prep = ArabertPreprocessor(model_name="bert-base-arabert", keep_emojis=False)


#Load Model
from transformers import EncoderDecoderModel, AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("tareknaous/bert2bert-empathetic-response-msa")
modell = EncoderDecoderModel.from_pretrained("tareknaous/bert2bert-empathetic-response-msa")

modell.to('cpu')
modell.eval()
print("done")
```

```python
#Function to generater response/post-process
def generate_response(text):
  text_clean = arabert_prep.preprocess(text)
  inputs = tokenizer.encode_plus(text_clean,return_tensors='pt')
  outputs = modell.generate(input_ids = inputs.input_ids.to('cpu'),
                   attention_mask = inputs.attention_mask.to('cpu'),
                   num_beams=1,
                   do_sample = True,
                   min_length=10,
                   top_k = 50,
                   temperature = 1,
                   length_penalty =2)
  preds = tokenizer.batch_decode(outputs)
  response = str(preds)
  response = response.replace("\'", '')
  response = response.replace("[[CLS]", '')
  response = response.replace("[SEP]]", '')
  response = str(arabert_prep.desegment(response))
  return response

```

```python
answers = {

     "greeting":["أهلا بك","مرحبا، كيف يمكنني مساعدتك","صباح الخير","أهلا، ماذا تحتاج؟"],
    "bye":["إلى اللقاء، أراك قريبا","إلى اللقاء","مع السلامة","وداعا","مع السلامة، أراك لاحقاً"]
}
def generate_prompt(user_input):
    prompt=""

    if user_input in ['أهلا'] :
        prompt += "CHATBOT:" +random.choice(answers["greeting"]) + "\n"

    elif user_input in ['إلى اللقاء']:
        prompt +="CHATBOT: "+ random.choice(answers["Bye"])+"\n"
    else:

        prompt +="CHATBOT: "+generate_response(user_input)
    return prompt
```

```python
user_input= input(" ME : ")
while user_input!='exit':
  print(generate_prompt(user_input))
  user_input= input(" ME : ")
```
